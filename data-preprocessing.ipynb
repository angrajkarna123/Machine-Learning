# Data Preprocessing

### Importing Libraries

```python
import pandas as pd
import numpy as np
```

### Loading Data

```python
data = pd.read_csv("../input/melbourne-housing-snapshot/melb_data.csv")
```

```python
data.head()
```

```python
data.columns
```

```python
data.describe()
```

```python
data.describe(include='object')
```

### Finding Categorical columns

```python
cat = (data.dtypes == 'object')
object_cols = list(cat[cat].index)
```

```python
print("Categorical variables are:")
print(object_cols)
```

```python
cat_data = data[object_cols]
```

```python
cat_data.head()
```

**This is separated Categorical data**

## Converting Categorical Data to Numeric Data

## Label Encoding

* I am applying Label Encoding to the ***Regionname*** column

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

cat_data['Regionname'] = label_encoder.fit_transform(cat_data['Regionname'])
```

```python
cat_data.head()
```

## One-Hot Encoding

I am applying One-Hot Encoding to the ***Method*** Column

```python
cat_data = pd.get_dummies(cat_data,columns=['Method'],prefix = ['cat'])
```

```python
cat_data.head()
```

### Replacing Categorical values with Numeric 

```python
cat_data['Type'].unique()
```

```python
type = {'h':0,'u':1,'t':2}
cat_data = cat_data.replace({'Type':type})
```

```python
cat_data.head()
```

## Droping Less important Columns

```python
drop = ['Suburb','Address','SellerG','Date','CouncilArea']

cat_data = cat_data.drop(drop, axis=1)
```

### Categorical Data which is converted to Numerical Data

```python
cat_data.head()
```

## Numeric Data

```python
num_data = data.select_dtypes(exclude=['object'])
```

```python
num_data.head()
```

## Concatinating Numeric and Converted Categorical Data

```python
new_data = pd.concat([num_data,cat_data],axis=1)
```

```python
new_data.head()
```

## Finding Columns With Null Values

```python
print(new_data.isnull().sum())
```

Columns ***BuildingArea*** , ***YearBuilt*** and ***Car*** has Null Values.

So I am droping These Columns.

## Droping Columns having Null Values

```python
new_data = new_data.drop(['BuildingArea','YearBuilt','Car'],axis=1)
```

```python
new_data.head()
```

```python
print(new_data.isnull().sum())
```

## Scaling Data


```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()

data_values = new_data.values

data_scaled = scaler.fit_transform(data_values)

normalized_data = pd.DataFrame(data_scaled)
```

```python
normalized_data.head()
```

```python
normalized_data.describe()
```

## Data Standardization

```python
from sklearn.preprocessing import StandardScaler

std_scaler = StandardScaler()

data_values = new_data.values

std_data = std_scaler.fit_transform(data_values)

standardized_data = pd.DataFrame(std_data)

standardized_data.head()
```

```python
standardized_data.describe()
```
